```python
# %% [markdown]
# # TP - Implémentation de LeNet-5 pour la reconnaissance de caractères amazighs (AMHCD)
# ## Architecture LeNet-5 reprogrammée pour la base AMHCD

# %%
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.utils import to_categorical
import cv2
import os
from PIL import Image
import warnings
warnings.filterwarnings('ignore')

# %%
# Configuration
IMG_SIZE = 64
BATCH_SIZE = 32
EPOCHS = 50
NUM_CLASSES = 33  # 33 caractères amazighs
TEST_SIZE = 0.2   # 20% pour le test
RANDOM_STATE = 42

# %%
# Chemin du dataset
dataset_path = "/kaggle/input/amazigh-handwritten-character-database-amhcd/AMHCD_64/AMHCD_64"
print("Dossiers disponibles:", os.listdir(dataset_path))

# %%
def load_and_split_dataset(dataset_path, test_size=0.2, random_state=42):
    images = []
    labels = []
    class_names = []
    
    # Lister les dossiers de classes (caractères amazighs)
    class_folders = sorted([f for f in os.listdir(dataset_path) 
                          if os.path.isdir(os.path.join(dataset_path, f))])
    
    print(f"Nombre de classes détectées: {len(class_folders)}")
    print("Classes:", class_folders)
    
    for class_idx, class_folder in enumerate(class_folders):
        class_path = os.path.join(dataset_path, class_folder)
        class_names.append(class_folder)
        
        # Compter les images par classe
        image_count = 0
        for img_file in os.listdir(class_path):
            if img_file.endswith(('.png', '.jpg', '.jpeg')):
                img_path = os.path.join(class_path, img_file)
                
                # Lecture de l'image
                img = cv2.imread(img_path)
                if img is not None:
                    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
                    
                    images.append(img)
                    labels.append(class_idx)
                    image_count += 1
        
        print(f"Classe {class_folder}: {image_count} images")
    
    # Conversion en arrays numpy
    images = np.array(images)
    labels = np.array(labels)
    
    # Split train/test
    X_train, X_test, y_train, y_test = train_test_split(
        images, labels, 
        test_size=test_size, 
        random_state=random_state,
        stratify=labels  # Même distribution dans train et test
    )
    
    return X_train, X_test, y_train, y_test, class_names

# %%
# Chargement et split des données
print("Chargement des données...")
X_train, X_test, y_train, y_test, class_names = load_and_split_dataset(
    dataset_path, test_size=TEST_SIZE, random_state=RANDOM_STATE
)

print(f"\nForme X_train: {X_train.shape}")
print(f"Forme y_train: {y_train.shape}")
print(f"Forme X_test: {X_test.shape}")
print(f"Forme y_test: {y_test.shape}")

# %%
# Normalisation des images
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# One-hot encoding des labels
y_train_categorical = to_categorical(y_train, NUM_CLASSES)
y_test_categorical = to_categorical(y_test, NUM_CLASSES)

# %%
# Visualisation de quelques exemples avec les noms des classes
plt.figure(figsize=(15, 10))
for i in range(15):
    plt.subplot(3, 5, i+1)
    plt.imshow(X_train[i])
    plt.title(f'{class_names[y_train[i]]}\n(Classe: {y_train[i]})', fontsize=9)
    plt.axis('off')
plt.tight_layout()
plt.show()

# %%
# Distribution des classes
plt.figure(figsize=(16, 6))

plt.subplot(1, 2, 1)
train_counts = np.bincount(y_train)
plt.bar(range(len(train_counts)), train_counts, alpha=0.7)
plt.xlabel('Classe')
plt.ylabel('Nombre d\'échantillons')
plt.title('Distribution des classes - Train')
plt.xticks(range(len(class_names)), class_names, rotation=90)

plt.subplot(1, 2, 2)
test_counts = np.bincount(y_test)
plt.bar(range(len(test_counts)), test_counts, alpha=0.7, color='orange')
plt.xlabel('Classe')
plt.ylabel('Nombre d\'échantillons')
plt.title('Distribution des classes - Test')
plt.xticks(range(len(class_names)), class_names, rotation=90)

plt.tight_layout()
plt.show()

# %%
# Architecture LeNet-5 modifiée pour AMHCD
def create_lenet5_amhcd_model(input_shape=(64, 64, 3), num_classes=33):
    model = keras.Sequential([
        # Couche de convolution 1
        layers.Conv2D(32, (5, 5), activation='relu', input_shape=input_shape, padding='same'),
        layers.BatchNormalization(),
        layers.AveragePooling2D((2, 2)),
        layers.Dropout(0.25),
        
        # Couche de convolution 2
        layers.Conv2D(64, (5, 5), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.AveragePooling2D((2, 2)),
        layers.Dropout(0.25),
        
        # Couche de convolution 3
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.AveragePooling2D((2, 2)),
        layers.Dropout(0.25),
        
        # Aplatissage
        layers.Flatten(),
        
        # Couches fully connected
        layers.Dense(256, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        
        layers.Dense(128, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        
        layers.Dense(num_classes, activation='softmax')
    ])
    
    return model

# %%
# Création du modèle
model = create_lenet5_amhcd_model()
model.summary()

# %%
# Compilation du modèle
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy', 'precision', 'recall']
)

# %%
# Callbacks
early_stopping = keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=15,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,
    patience=8,
    min_lr=0.00001,
    verbose=1
)

checkpoint = keras.callbacks.ModelCheckpoint(
    'best_lenet5_amhcd_model.h5',
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)

# %%
# Entraînement du modèle
print("Début de l'entraînement...")
history = model.fit(
    X_train, y_train_categorical,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    validation_data=(X_test, y_test_categorical),
    callbacks=[early_stopping, reduce_lr, checkpoint],
    verbose=1
)

# %%
# Évaluation du modèle
print("Évaluation du modèle...")
test_loss, test_acc, test_precision, test_recall = model.evaluate(
    X_test, y_test_categorical, verbose=0
)

print(f"\nRésultats sur le test set:")
print(f"Accuracy: {test_acc:.4f}")
print(f"Loss: {test_loss:.4f}")
print(f"Precision: {test_precision:.4f}")
print(f"Recall: {test_recall:.4f}")

# %%
# Courbes d'apprentissage
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Accuracy
axes[0, 0].plot(history.history['accuracy'], label='Train Accuracy')
axes[0, 0].plot(history.history['val_accuracy'], label='Validation Accuracy')
axes[0, 0].set_title('Accuracy du modèle')
axes[0, 0].set_xlabel('Epochs')
axes[0, 0].set_ylabel('Accuracy')
axes[0, 0].legend()
axes[0, 0].grid(True)

# Loss
axes[0, 1].plot(history.history['loss'], label='Train Loss')
axes[0, 1].plot(history.history['val_loss'], label='Validation Loss')
axes[0, 1].set_title('Loss du modèle')
axes[0, 1].set_xlabel('Epochs')
axes[0, 1].set_ylabel('Loss')
axes[0, 1].legend()
axes[0, 1].grid(True)

# Precision
axes[1, 0].plot(history.history['precision'], label='Train Precision')
axes[1, 0].plot(history.history['val_precision'], label='Validation Precision')
axes[1, 0].set_title('Precision du modèle')
axes[1, 0].set_xlabel('Epochs')
axes[1, 0].set_ylabel('Precision')
axes[1, 0].legend()
axes[1, 0].grid(True)

# Recall
axes[1, 1].plot(history.history['recall'], label='Train Recall')
axes[1, 1].plot(history.history['val_recall'], label='Validation Recall')
axes[1, 1].set_title('Recall du modèle')
axes[1, 1].set_xlabel('Epochs')
axes[1, 1].set_ylabel('Recall')
axes[1, 1].legend()
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()

# %%
# Prédictions
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# %%
# Matrice de confusion
plt.figure(figsize=(16, 14))
cm = confusion_matrix(y_test, y_pred_classes)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=class_names, 
            yticklabels=class_names,
            annot_kws={'size': 8})
plt.title('Matrice de confusion - LeNet-5 sur AMHCD', fontsize=16)
plt.xlabel('Prédictions', fontsize=12)
plt.ylabel('Vraies labels', fontsize=12)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()

# %%
# Rapport de classification détaillé
print("Rapport de classification détaillé:")
print(classification_report(y_test, y_pred_classes, 
                          target_names=class_names))

# %%
# Accuracy par classe
class_accuracy = []
for i in range(NUM_CLASSES):
    class_mask = (y_test == i)
    if np.sum(class_mask) > 0:
        class_acc = np.mean(y_pred_classes[class_mask] == i)
        class_accuracy.append(class_acc)
    else:
        class_accuracy.append(0)

# Plot accuracy par classe
plt.figure(figsize=(16, 6))
plt.bar(range(NUM_CLASSES), class_accuracy)
plt.xlabel('Classe')
plt.ylabel('Accuracy')
plt.title('Accuracy par classe')
plt.xticks(range(NUM_CLASSES), class_names, rotation=90)
plt.ylim(0, 1)
plt.grid(axis='y', alpha=0.3)
plt.show()

# %%
# Visualisation des prédictions (bonnes et mauvaises)
correct_indices = np.where(y_pred_classes == y_test)[0]
incorrect_indices = np.where(y_pred_classes != y_test)[0]

print(f"Nombre de prédictions correctes: {len(correct_indices)}")
print(f"Nombre d'erreurs: {len(incorrect_indices)}")
print(f"Accuracy globale: {len(correct_indices)/len(y_test):.4f}")

# %%
# Visualisation de quelques bonnes prédictions
plt.figure(figsize=(15, 12))
for i, idx in enumerate(correct_indices[:12]):
    plt.subplot(3, 4, i+1)
    plt.imshow(X_test[idx])
    
    true_label = y_test[idx]
    pred_label = y_pred_classes[idx]
    confidence = np.max(y_pred[idx])
    
    plt.title(f'True: {class_names[true_label]}\nPred: {class_names[pred_label]}\nConf: {confidence:.2f}', 
              color='green', fontsize=8)
    plt.axis('off')

plt.suptitle('Bonnes prédictions', fontsize=16)
plt.tight_layout()
plt.show()

# %%
# Visualisation de quelques erreurs
plt.figure(figsize=(15, 12))
for i, idx in enumerate(incorrect_indices[:12]):
    plt.subplot(3, 4, i+1)
    plt.imshow(X_test[idx])
    
    true_label = y_test[idx]
    pred_label = y_pred_classes[idx]
    confidence = np.max(y_pred[idx])
    
    plt.title(f'True: {class_names[true_label]}\nPred: {class_names[pred_label]}\nConf: {confidence:.2f}', 
              color='red', fontsize=8)
    plt.axis('off')

plt.suptitle('Erreurs de prédiction', fontsize=16)
plt.tight_layout()
plt.show()

# %%
# Sauvegarde du modèle final
model.save('lenet5_amhcd_final_model.h5')
print("Modèle final sauvegardé!")

# %%
# Analyse des classes les plus difficiles
class_errors = []
for i in range(NUM_CLASSES):
    class_mask = (y_test == i)
    if np.sum(class_mask) > 0:
        error_rate = 1 - class_accuracy[i]
        class_errors.append((class_names[i], error_rate, np.sum(class_mask)))

# Trier par taux d'erreur
class_errors.sort(key=lambda x: x[1], reverse=True)

print("\nClasses les plus difficiles (taux d'erreur):")
for class_name, error_rate, count in class_errors[:10]:
    print(f"{class_name}: {error_rate:.3f} ({count} échantillons)")

# %%
# Feature maps visualization (optionnel)
def visualize_feature_maps(model, image, layer_names):
    """Visualise les feature maps des couches convolutionnelles"""
    layer_outputs = [layer.output for layer in model.layers if layer.name in layer_names]
    activation_model = keras.models.Model(inputs=model.input, outputs=layer_outputs)
    
    activations = activation_model.predict(np.expand_dims(image, axis=0))
    
    return activations

# Exemple de visualisation pour une image
sample_image = X_test[0]
layer_names = ['conv2d', 'conv2d_1', 'conv2d_2']  # Noms des couches convolutionnelles
activations = visualize_feature_maps(model, sample_image, layer_names)

# %%
print("TP terminé avec succès!")
```
